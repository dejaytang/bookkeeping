```python
from management.transaction import Income, Expense
from management.book import Book
from analysis.budget import BudgetAnalysis
from analysis.summary import SummaryAnalysis

import tkinter as tk
from tkinter import simpledialog
```

```python
def get_user_input(prompt):
    # Create a root window
    root = tk.Tk()
    # Hide the root window
    root.withdraw()

    # Open a simple dialog to get user input
    user_input = simpledialog.askstring(title="Input Required", prompt=prompt)

    # The window will remain open until the user provides input and closes the dialog box
    return user_input

def main():
    while True:
        prompt = """Please select from the below options (use numbers please):
        1. Manage transactions
        2. Show analysis
        3. Quit"""

        response = int(get_user_input(prompt))
        if response == 3:
            break
        elif response == 1:
            book = Book()
            prompt = """Please select from the below options (use numbers please):
            1. Add a new transaction
            2. Delete a transaction
            3. Search a transaction
            4. Show the existing transactions
            5. Quit"""
            response = int(get_user_input(prompt))
            if response == 1:
                type = get_user_input("Please specify the type of the transaction: income or expense")
                if type == "income":
                    response = get_user_input(f"""Please input the amount, description, and label (seperated by comma)
                    Available label options: salary income, passive income, capital gains, other""")
                    items = response.split(',')
                    items = [item.strip() for item in items]
                    amount = items[0]
                    desc = items[1]
                    label = items[2]
                    book.addTran(Income(amount, desc, label))
                elif type == "expense":
                    response = get_user_input(f"""Please input the amount, description, and label (seperated by comma)
                    Available label options: utility, grocery, transportation, entertainment, healthcare, other""")
                    items = response.split(',')
                    items = [item.strip() for item in items]
                    amount = items[0]
                    desc = items[1]
                    label = items[2]
                    book.addTran(Expense(amount, desc, label))
            elif response == 2:
                type = get_user_input("Please specify the type of the transaction: income or expense")
                index = get_user_input("Please specify the index of the to be deleted transaction")
                book.removeTran(type, index)
            elif response == 3:
                type = get_user_input("Please specify the type of the transaction: income or expense")
                keyword = get_user_input("Please specify the keyword of the to be searched transaction")
                book.searchTran(type, keyword)
            elif response == 4:
                book.displayBook()
            else:
                break
        elif response == 2:
            type = get_user_input("Please specify the type of the analysis: budget or overall")
            if type == "budget":
                ba = BudgetAnalysis()
                b_type = get_user_input("""Please specify how the budget is calculated:
                1. Amount
                2. Percentage of overall income""")
                budget = int(get_user_input("Please enter a budget or percentage (0 to 100):"))
                ba.setBudget(budget, b_type)
                ba.overBudgetExpenses()
                ba.displayOverBudgetTrend()
            elif type == 'overall':
                sa = SummaryAnalysis()
                response = int(get_user_input(
                    """Please select the analysis:
                       1. Check current balance
                       2. Show the trend of expenses
                       3. Check the distribution of labels"""))
                if response == 1:
                    sa.checkBalance()
                elif response == 2:
                    type = int(get_user_input("Please specify the type of graph to show: \n1. Line chart\n2. Bar chart"))
                    sa.displayTrend(type)
                elif response == 3:
                    type = int(get_user_input("""
                            Please specify the type of graph to show:
                            1. for pie chart
                            2. for donut chart
                            3. for ring chart
                            4. for half-pie chart"""))
                    transaction = get_user_input("Please select the transaction type that the analysis is about:\nincome or expense")
                    sa.displayByLabel(transaction, type)
                

if __name__ == "__main__":
    main()
```
